```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
res = pd.read_csv("/home/vs428/Documents/Uncertainty/all_cases_posneg_gpt4_output_race_noLRreasoning_v2.csv")
```

```python
fig, axs = plt.subplots(1,2, figsize=(12,5))
sns.barplot(res[res['positive'] == True], x="case_type", y="bayes_diff", hue="race", ax=axs[0]).set_title('Positive Lab')
sns.barplot(res[res['positive'] == False], x="case_type", y="bayes_diff", hue="race", ax=axs[1]).set_title('Negative Lab')
```

```python
fig, axs = plt.subplots(1,2, figsize=(12,5))
sns.barplot(res[(res['positive'] == True) & 
            (res['case_type'].isin(['CHF', "Pulmonary Embolism"]))], 
            x="case_type", y="bayes_diff", hue="race", ax=axs[0]).set_title('Positive Lab')
sns.barplot(res[(res['positive'] == False) & 
            (res['case_type'].isin(['CHF', "Pulmonary Embolism"]))], 
            x="case_type", y="bayes_diff", hue="race", ax=axs[1]).set_title('Negative Lab')
```

```python
fig, axs = plt.subplots(1,2, figsize=(12,5))
sns.barplot(res[(res['positive'] == True) & 
            (res['case_type'].isin(['ACS', "PNEUMONIA"]))], 
            x="case_type", y="bayes_diff", hue="race", ax=axs[0]).set_title('Positive Lab')
sns.barplot(res[(res['positive'] == False) & 
            (res['case_type'].isin(['ACS', "PNEUMONIA"]))], 
            x="case_type", y="bayes_diff", hue="race", ax=axs[1]).set_title('Negative Lab')
```
