# Sensitivity/Specificity

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
res = pd.read_csv("/home/vs428/Documents/Uncertainty/all_cases_posneg_gpt4_output_sensspec_v5.csv")
```

```python
res.loc[res['race'] == "African American", "sens"].describe(), res.loc[res['race'] == "White", "sens"].describe()
```

```python
res.loc[res['race'] == "African American", "spec"].describe(), res.loc[res['race'] == "White", "spec"].describe()
```

```python
fig, axs = plt.subplots(2,2, figsize=(12,5))
sns.barplot(res[res['positive'] == True], x="case_type", y="sens", hue="race", ax=axs[0][0]).set_title('Sensitivity')
sns.barplot(res[res['positive'] == True], x="case_type", y="spec", hue="race", ax=axs[0][1]).set_title('Specificity')

sns.barplot(res[res['positive'] == False], x="case_type", y="sens", hue="race", ax=axs[1][0]).set_title('Sensitivity')
sns.barplot(res[res['positive'] == False], x="case_type", y="spec", hue="race", ax=axs[1][1]).set_title('Specificity')
```
