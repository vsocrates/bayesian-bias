# Clean up unicode in Text

```python
import pandas as pd
import csv
```

```python
data = pd.read_csv("all_cases.csv", sep="|",  engine="c")
```

```python
data['case'] = data['case'].str.replace("“", '"')
data['case'] = data['case'].str.replace("”", '"')
data['case'] = data['case'].str.replace("’", "'")
data['case'] = data['case'].str.replace("½", "1/2")
data['case'] = data['case'].str.replace("–", "-")
```

```python
import traceback 

try:
    for x in data['lab_value_text'].tolist():
        x.encode('ascii')
except UnicodeEncodeError:
    print("it was not a ascii-encoded unicode string: ", x)
    print(traceback.format_exc())
```

```python
data.to_csv("all_cases.csv", sep="|")
```
