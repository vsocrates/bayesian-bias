```python
# thresholds for low, medium, and high risks of CAP based on validated risk score
# low (probability <2.5%), intermediate (2.5-20%), and high (>20%)
```

We want two data points from this study

1. The percentage difference from estimation to true, betweeen the physicians and the LLMs.
    1. The first one will be lumped into the pre-test data comparison with risk-score calculators data I get from Thomas/Don
2. We will figure out how many people changed their decisions for low, medium, and high CAP-risk groups, for both the LLMs and the physicians



```python
# for (2), we first need to split up the physician data into risk groups
def add_CAP_risk_levels(df, pretest_prob_colname, true_prob_colname):
    # for (2), we first need to split up the physician data into risk groups
    df["true_risk"] = np.where(df[true_prob_colname] < 2.5, "low", np.nan)
    df["true_risk"] = np.where((df[true_prob_colname] >= 2.5) & (df[true_prob_colname] <= 20) , "medium", df["true_risk"])
    df["true_risk"] = np.where((df[true_prob_colname] > 20), "high", df["true_risk"])    
    
    df["estimate_risk"] = np.where(df[pretest_prob_colname] < 2.5, "low", np.nan)
    df["estimate_risk"] = np.where((df[pretest_prob_colname] >= 2.5) & (df[pretest_prob_colname] <= 20) , "medium", df["estimate_risk"])
    df["estimate_risk"] = np.where((df[pretest_prob_colname] > 20), "high", df["estimate_risk"])

    return df 
    
physician_data = add_CAP_risk_levels(physician_data, "scenariopercent", "PercentPneumonia")
```

```python
physician_data["true_risk"].value_counts(), physician_data["estimate_risk"].value_counts() 
```

There's a number of situations that can occur for (2). 

- Initial estimate and true prob could be in different risk categories, and then the individual changes their decision.
- Initial estimate and true prob could be in different risk categories, and the individual keeps their decision. 
- Initial estimate and true prob could be in same risk categories, and the individual changes their decision.
- Initial estimate and true prob could be in same risk categories, and the individual keeps their decision.

*The question we want to answer is, when the true probability is in a different risk group to the estimated probability, what does the LLM do? How about the physician? How do they differ?*

```python
# let's try one
different_risks = physician_data[physician_data['true_risk'] != physician_data['estimate_risk']]
```

```python
# ask what AB, BC, and Z are, but for now, drop em. 
display(different_risks["scenariocategory"].value_counts(), different_risks["scenariorevised"].value_counts())

different_risks = different_risks[(different_risks['scenariocategory'].isin(['A', "B", "C"])) &
                (different_risks['scenariorevised'].isin(['A', "B", "C"]))]
```

```python
different_risks_low = different_risks[different_risks['true_risk'] == "low"]
different_risks_medium = different_risks[different_risks['true_risk'] == "medium"]
different_risks_high = different_risks[different_risks['true_risk'] == "high"]
```

```python
different_risks_counts = different_risks[['scenariocategory', "scenariorevised"]].value_counts(normalize=True).sort_index()
```

```python
display(different_risks_counts)
different_risks_counts.loc[('A', "A")] + different_risks_counts.loc[('B', "B")] + different_risks_counts.loc[('C', "C")]
```

```python
# look at it by risk group: 
display(different_risks_low[['scenariocategory', "scenariorevised"]].value_counts(normalize=True).sort_index())
display(different_risks_medium[['scenariocategory', "scenariorevised"]].value_counts(normalize=True).sort_index())
display(different_risks_high[['scenariocategory', "scenariorevised"]].value_counts(normalize=True).sort_index())
```

```python
different_risks_medium_BtoA = different_risks_medium[(different_risks_medium['scenariocategory'] == "B") & (different_risks_medium['scenariorevised'] == "A")]
(different_risks_medium_BtoA["scenariopercent"] - different_risks_medium_BtoA["PercentPneumonia"]).describe()
# we see here that in all cases where the decision went from treat to rule-out, the initial estimate was much higher. 
```
